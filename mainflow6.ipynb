```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
from datetime import datetime
```

```python
df=pd.read_csv('disney_plus_titles.csv')
df.head()
```

```python
print(sorted(df["release_year"].unique()))
```

```python
df.head()
```

```python
df.loc[(df["type"]== 'Movie'), "category_name"] = 'M'
df.loc[(df["type"]== 'TV Show'), "category_name"] = 'TV'
df.head()
```

```python
df['year']= df['release_year']
yearly_counts= df.groupby('year')['category_name'].count()
yearly_counts.plot(kind='bar', xlabel='Year', ylabel='release', title='Total publish video per year')
plt.show()
```

```python
df.category_name.value_counts().plot(kind='bar', color=["salmon", "orchid"])
plt.title("M or tv")
plt.xlabel("release_year")
plt.ylabel("M=Movie, TV=Television")
```

```python
srn.countplot(x= 'category_name', data=df)
plt.title("tv mv")
plt.xlabel("M=Movie, TV=Television")
```

```python
df.hist(bins=100,grid=False, figsize=(20,15))
```

```python
#how many movies and hoe many tv shows
df.type.value_counts()
```

```python
plt.bar (df['release_year'], df['type'])
plt.title("Bar Chart")
# Setting the X and Y labels
plt.xlabel ('tear' )
plt.ylabel( 'type')
# Adding the legends
plt. show()
```

```python
pd.crosstab(df.type,df.release_year)
```

```python
srn.countplot(x= 'type', data=df, hue='release_year')
plt.title("tv")
plt.xlabel("M=Movie, TV=Television")
```
